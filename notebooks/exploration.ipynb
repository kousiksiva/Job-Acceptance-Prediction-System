```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("data/raw_job_data.csv")
df.head()
```

```python
print("Shape:", df.shape)
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
target_col = None
for col in df.columns:
    if "place" in col.lower() or "status" in col.lower():
        target_col = col
        break

print("Target column:", target_col)
```

```python
sns.countplot(data=df, x=target_col)
plt.title("Placement Status Distribution")
plt.show()
```

```python
sns.boxplot(data=df, x=target_col, y="technical_score")
plt.title("Technical Score vs Placement")
plt.show()

sns.boxplot(data=df, x=target_col, y="communication_score")
plt.title("Communication Score vs Placement")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="degree_specialization", hue=target_col)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.show()
```

```python
## Key Observations
- Higher technical and communication scores increase placement probability.
- Degree specialization impacts placement.
- Dataset contains real-world variability and noise.
```
